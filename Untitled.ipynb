```python
import pandas as pd
import plotly.express as px

df = pd.read_csv(r"data\csv\centerline.csv")

fig = px.line_3d(df, x="x", y="y", z="z", markers=True)

fig.update_layout(
    scene_aspectmode="data",
    scene_camera=dict(
        center=dict(x=0, y=0, z=0),
        eye=dict(x=1.0, y=1.0, z=0.6)
    )
)

fig.show()
```
